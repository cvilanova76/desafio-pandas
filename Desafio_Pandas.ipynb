Quantas pessoas morreram e quantas sobreviveram?

````python
total_sobreviventes = titanic.query("survived == 'yes'").count()
total_sobreviventes
````

Outra maneira de contar

````python
titanic['survived'].value_counts()
````

Quantos eram da primeira, da segunda e da terceira classe

````python
titanic['class'].value_counts()
````

Qual a porcentagem de sobreviventes na primeira classe?

Primeiro classificar a informação de sobrevivente ou não por classe
```



````python
sobreviventes_por_classe = titanic.groupby(['class', 'survived']).count()
sobreviventes_por_classe
````

Depois calcular a porcentagem de sobreviventes na primeira classe

````python
total_sobreviventes_1classe = titanic[(titanic['survived'] == 'yes') & (titanic['class'] == '1st class')].count()
total_sobreviventes_1classe
````

````python
porcentagem_1classe = (total_sobreviventes_1classe)*100/(total_sobreviventes)
porcentagem_1classe
````

Qual a porcentagem de mortos da terceira classe

````python
total_sobreviventes_3classe = titanic[(titanic['survived'] == 'yes') & (titanic['class'] == '3rd class')].count()
total_sobreviventes_3classe
````

````python
porcentagem_3classe = total_sobreviventes_3classe*100/total_sobreviventes
porcentagem_3classe
````

**Desafio 5**
Forbes

````python
from pydataset import data
forbes = data("Forbes2000")
forbes
````

Qual a empresa mais valiosa da lista Forbes2000 (parâmetro = marketvalue)

````python
maximo = forbes["marketvalue"].max()
maximo
````

````python
mais_valiosa = forbes[forbes["marketvalue"] == maximo]
mais_valiosa
````

Construa uma tabela com as TOP 10 empresas mais lucrativas da lista (parâmetro = profits)

````python
top10 = forbes.nlargest(10, "profits")
top10
````

Qual a média de valores de empresa das cinco categorias mais valiosas?

````python
top5 = forbes.nlargest(5, "marketvalue")
top5['category'].unique()
````

````python
top5.groupby('category').mean().round(2)
````
