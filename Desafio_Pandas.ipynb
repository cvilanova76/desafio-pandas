**Desafio 3**
Titanic

```python
from pydataset import data
titanic = data("titanic")
```

visualizando os dados do dataset

```python
titanic.head()
```

```python
titanic.shape
```

```python
titanic.tail()
```

Quantas pessoas morreram e quantas sobreviveram?

```python
titanic.query("survived == 'no'")
```

```python
total_sobreviventes = titanic.query("survived == 'yes'").count()
total_sobreviventes
```

Outra maneira de contar

```python
titanic['survived'].value_counts()
```

Quantos eram da primeira, da segunda e da terceira classe

```python
titanic['class'].value_counts()
```

Qual a porcentagem de sobreviventes na primeira classe?

```python
sobreviventes = titanic.value_counts([(titanic['class'] == "1st class") & (titanic['survived'] == "yes")])
total = titanic.value_counts([(titanic['class'] == "1st class")])
calculo = (sobreviventes[1] / total[1]) * 100
print(f"Porcentagem de sobreviventes na primeira classe: {calculo:.2f}%")
```

Qual a porcentagem de mortos da terceira classe

```python
morreram = titanic.value_counts([(titanic['class'] == "3rd class") & (titanic['survived'] == "no")])
total = titanic.value_counts([(titanic['class'] == "3rd class")])
calculo = (morreram[1] / total[0]) * 100
print(f"Porcentagem de mortos na terceira classe: {calculo:.2f}%")
```

Criar a coluna com o total de gols de cada partida

```python
bundesliga["Total_Goals"] = bundesliga["AwayGoals"] + bundesliga["HomeGoals"]
bundesliga
```

Classificar a tabela a partir da coluna de total de gols, do maior para menor, para mostrar a partida com mais gols na história da Bundesliga

**Desafio 5**
Forbes

```python
from pydataset import data
forbes = data("Forbes2000")
forbes
```

Qual a empresa mais valiosa da lista Forbes2000 (parâmetro = marketvalue)

```python
maximo = forbes["marketvalue"].max()
maximo
```

```python
mais_valiosa = forbes[forbes["marketvalue"] == maximo]
mais_valiosa
```

Construa uma tabela com as TOP 10 empresas mais lucrativas da lista (parâmetro = profits)

```python
top10 = forbes.nlargest(10, "profits")
top10
```

Qual a média de valores de empresa das cinco categorias mais valiosas?

```python
top5 = forbes.nlargest(5, "marketvalue")
top5['category'].unique()
```

```python
mkt_v_somados = forbes[['marketvalue','category']].groupby('category').sum().nlargest(5,'marketvalue')
mkt_v_somados
```

```python
top5.groupby('category').mean(numeric_only=True).round(2)
```

```python
med_categ = mkt_v_somados.mean()
print("Média de valores das cinco categorias mais valiosas: ", med_categ)
```
